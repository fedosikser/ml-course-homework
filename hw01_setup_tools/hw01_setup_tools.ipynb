```python
import numpy as np

np.random.seed(42)

X = np.random.randn(100, 5)

print("X.shape:", X.shape)
print("X.dtype:", X.dtype)
print("First 5 rows:\n", X[:5])

# Задание 3.2
print("\n" + "="*50)
print("СТАТИСТИКИ МАТРИЦЫ X (100×5)")
print("="*50)

# Среднее и стандартное отклонение по всей матрице
mean_all = X.mean()
std_all = X.std()
print(f"Среднее по всей матрице: {mean_all:.4f}")
print(f"Стандартное отклонение по всей матрице: {std_all:.4f}")

# Среднее по каждому столбцу (axis=0)
mean_by_column = X.mean(axis=0)
print("\nСредние значения по столбцам:")
for i, mean_val in enumerate(mean_by_column):
    print(f"  Столбец {i}: {mean_val:.4f}")

# Стандартное отклонение по каждому столбцу
std_by_column = X.std(axis=0)
print("\nСтандартные отклонения по столбцам:")
for i, std_val in enumerate(std_by_column):
    print(f"  Столбец {i}: {std_val:.4f}")

# Минимумы по каждому столбцу
min_by_column = X.min(axis=0)
print("\nМинимальные значения по столбцам:")
for i, min_val in enumerate(min_by_column):
    print(f"  Столбец {i}: {min_val:.4f}")

# Максимумы по каждому столбцу
max_by_column = X.max(axis=0)
print("\nМаксимальные значения по столбцам:")
for i, max_val in enumerate(max_by_column):
    print(f"  Столбец {i}: {max_val:.4f}")

# Для более компактного вывода можно также вывести всё в виде таблицы:
print("\n" + "="*50)
print("СВОДНАЯ ТАБЛИЦА СТАТИСТИК")
print("="*50)
print(f"{'Столбец':<10} {'Среднее':<12} {'Стд.откл.':<12} {'Минимум':<12} {'Максимум':<12}")
print("-"*58)
for i in range(5):
    print(f"{i:<10} {mean_by_column[i]:<12.4f} {std_by_column[i]:<12.4f} "
          f"{min_by_column[i]:<12.4f} {max_by_column[i]:<12.4f}")
```
